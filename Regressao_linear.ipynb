## Regressao Linear

**Vairaveis:**

- Dua variaveis numericas. Existe uma relaçao matemática entre estas duas variaveis?
- Se existe, como posso medir sua força?
- Poderia usar essa relaçao para fazer previsoes?

**Grafico de dispersao**
- é a representaçao gráfica principal para regressao Linear
- O Eixo Y Usamos como a variavel de resposta ou a variavel dependente (é aquilo que eu quero prever).
- Eixo X é a variavel Explanatoria ou Independente. (É aquela que explica e que usamos para prever).


**Correlaçao:**
- Mostra a força e a direçao da relaçao entre variaveis.
    - Força e direçao
- Pode ser um valor real qualqque entre -1 e 1
- A correlao entre A~B é igual a B~A

No grafico de disperçao uma correlaçao forte ela vai estar com os dados agrupados e alinhados, existe uma relaçao matematica forte. Ja uma correlaçao fraca vao ter os dados dispersos e desalinhados.

**Coeficiente de Determinaçao (R²):**
- Mostra o quanto o modelo consegue explicar os valores 
- Quanto maior mais explicativo é
- O restante da variabilidade está em variáveis nao incluidas no modelo
- Varia entre 0 e 1
- Calcula-se com o quadrado do coeficiente de correlaçao

**Residuos**

O valor do ponto do dado que temos menos o mesmo valor no ponto da linha de regressao.
- Eles indicam o erro de previsão do modelo e ajudam a avaliar sua precisão e adequação aos dados.
- Se o ponto estiver coincidindo com a linha o residuo sera 0.



**Extrapolação:**
    
É a estimativa de valores fora do intervalo dos dados observados, usando a relação linear encontrada. Ela é arriscada, pois pode não refletir corretamente o comportamento real dos dados além do intervalo estudado.

**Outliers**

São valores que se distanciam significativamente da maioria dos dados em um conjunto, podendo influenciar a média, a variância e até a eficácia de um modelo de regressão. Na análise de regressão, outliers podem distorcer a linha de melhor ajuste e reduzir a precisão das previsões, por isso é importante identificá-los e avaliar seu impacto.

**CONDIÇOES PARA REGRESSAO LINEAR**

- **Correlaçao:** deve ser de moderada a forte
- Coefeficiente de determinaçao R²: 
    - Entre 0 e 0.3 (Ruim): Nao é indicado usar regressao Linear
    - Entre eles: Vale analise para decidir
    - Acima de 0,7 (ótimo): Pode-se usar.
- Residuos: Próximos de uma distribuiçao normal
    - Histograma
    - Diagrama de normailidade
    - Teste de Shapiro Wilk

**Regressao Linear Simples e Múltipla**

- **Simples:**
    - Tem uma variavel explanatoria para prever uma variável dependente
    - Y~X

- **Multipla:**
    - Duas ou mais variáveis explanatorias para prever a variável dependente
    - Y~ X1 + X1 + Xn

***Nao necessáriamente a variavel multipla vai gerar um melhor modelo. Isso vai depender dos dados

- Analisar cada variável independente com Y individualmente
- Gerar gráficos de dispersao individuais
- Buscar redundancias (mesmo efeitos de x sobre y)

**Coeficiente de Determinaçao (R²):**
- R² é o percentual de variacao da variavel de resposta que é explicada pelo modelo
- Quando se colocam mais variáveis no modelo, a tendencia é que o R² aumente, mesmo que a adiaçao da var nao aumente a precisao do modelo
- Solução: R²ajustado. Para resolver isso, existe o R² ajustado, que ajusta o valor do R² de acordo com o número de variáveis adicionadas. O R² ajustado só aumenta se a nova variável realmente ajuda o modelo e melhora a explicação dos dados de maneira significativa. Caso contrário, ele pode até diminuir, indicando que aquela variável não foi tão útil assim.



**Colinearidade**

Ocorre quando duas ou mais variáveis independentes (ou preditoras) estão fortemente relacionadas entre si. Isso quer dizer que uma dessas variáveis pode ser “prevista” usando outra, ou seja, elas contêm quase a mesma informação.

- Por que colinearidade é um problema?\
    - Dificulta a interpretação: Fica difícil entender a importância real de cada variável no modelo, já que elas estão “brigando” para representar a mesma coisa.\
    - Afeta a precisão das estimativas: O modelo pode fazer previsões menos confiáveis, pois a presença de variáveis colineares aumenta a incerteza das estimativas.

- Como evitar a colinearidade?
    - Usar uma análise de correlação: para verificar se variáveis estão fortemente relacionadas (correlação alta).
    - Remover variáveis redundantes: manter apenas uma das variáveis fortemente correlacionadas.
    - Usar métodos de regularização, como Ridge ou Lasso, que ajudam a lidar com colinearidade.

**Parcimônia**

É o princípio de que o modelo deve ser o mais simples possível, mas sem perder sua capacidade de explicar os dados. Em outras palavras, devemos incluir apenas as variáveis que realmente ajudam a melhorar a previsão ou explicação dos dados e evitar variáveis desnecessárias.

- Por que parcimônia é importante?
    - Facilidade de interpretação: Um modelo com menos variáveis é mais fácil de entender e explicar.
    - Reduz o risco de overfitting: Um modelo muito complexo pode “aprender” o ruído dos dados, ou seja, capturar padrões que são apenas coincidências, não relações reais.
    - Maior generalização: Modelos simples geralmente funcionam melhor em dados novos, enquanto modelos complexos podem funcionar bem apenas nos dados em que foram treinados.

- Como alcançar parcimônia?
    - Analisar a relevância das variáveis: Usar técnicas como o R² ajustado ou a seleção de variáveis para incluir apenas as variáveis mais significativas.
    - Preferir modelos mais simples: Quando dois modelos têm desempenhos similares, é melhor escolher o mais simples.
    - Eliminar variáveis redundantes ou irrelevantes: Usar técnicas como testes de significância estatística e métodos de seleção de variáveis, como Backward Selection ou Forward Selection



**RESUMINDO:**

- **Colinearidade:** Problema quando duas variáveis contêm a mesma informação. Evitar colinearidade melhora a interpretação e a precisão do modelo.
- **Parcimônia:** Princípio de simplicidade – usar o menor número possível de variáveis que ainda expliquem bem os dados. Isso ajuda a criar modelos mais robustos e interpretáveis.

**REQUISITOS BÁSICOS:**
- Linearidade entre a variável dependente e as variáveis independentes
- Pouca ou nenhuma Colinearidade


**Resíduos**

- Assim como na simples, esperamos que seja Próximos a distribuiçao normal
- Variancia constante em relaçao a linha de melhor ajuste
- Independentes (sem padrao)

**Correlogramas**

Mostra todas essas correlações em uma única figura, geralmente como uma matriz. Em cada célula da matriz, ele mostra a correlação entre duas variáveis, de acordo com as seguintes convenções: Cores, Numeros e formas.

- Como interpretar um correlograma?
    - Identificar relações fortes
    - Identificar colinearidade
    - Descobrir padrões

****Correlacao nao é causa:** a correlação mede apenas a associação entre duas variáveis, mas não indica que uma variável faz a outra acontecer.
